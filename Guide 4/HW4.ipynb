## Computer Networking Study Guide

### End-to-End Protocols

1. What is the purpose of end-to-end protocols in computer networking?
* Los protocolos de extremo a extremo garantizan una comunicación confiable entre el origen y el destino en las redes informáticas. Gestionan la entrega de datos, el manejo de errores y el control de flujo para garantizar que los datos se transmitan correctamente de un sistema final a otro.
2. Explain the difference between connection-oriented and connectionless protocols.
* Orientado a la conexión: establece una conexión dedicada antes de la transferencia de datos, proporcionando confiabilidad a través de mecanismos como el protocolo de enlace y la retransmisión (por ejemplo, TCP).
* Sin conexión: Transmite datos sin acuerdo previo y la confiabilidad es manejada por capas superiores o no está garantizada en absoluto (por ejemplo, UDP).

3. What are the advantages and disadvantages of using end-to-end protocols?
* Ventajas: Entrega de datos confiable, detección y corrección de errores y control de flujo.
* Desventajas: Gastos generales debidos a la configuración y el mantenimiento de la conexión, potencialmente más lentos que los protocolos sin conexión.

4. How does the Internet Protocol (IP) provide end-to-end delivery?
* IP es un protocolo de capa de red responsable de enrutar paquetes a través de redes. Proporciona la base para la entrega de un extremo a otro al dirigir y reenviar paquetes a su destino.
5. What is the role of the Transport Layer in end-to-end protocols?
* La capa de transporte gestiona la comunicación de un extremo a otro y ofrece servicios como segmentación, reensamblaje, detección de errores y control de flujo.

### UDP

1. What is UDP and how does it differ from TCP?
* UDP es un protocolo de capa de transporte sin conexión que proporciona un conjunto mínimo de funciones para transmitir datos entre aplicaciones. Opera sin establecer una conexión antes de enviar datos y no garantiza la entrega ni el orden de los paquetes. La diferencia que tiene con TCP es que no se requiere configuración de conexión antes de la transmisión de datos, no hay garantía de entrega y los paquetes pueden llegar desordenados, menos gastos generales de protocolo en comparación con TCP, debido a su simplicidad, UDP es generalmente más rápido que TCP.
2. Explain the advantages and disadvantages of using UDP.
* Usar UDP tiene varias ventajas como la baja latencia, esto hace que la transmisión sea más rápida sin la sobrecarga de configuración de la conexión, la simplicidad, transmisión y multidifusión.
* Las desventajas de usar UDP es que no es confiable porque no detecta ni corrige ningún error y es posible que se pierdan paquetes, no tiene control de flujos por lo que no es apto para aplicaciones que requieren confiabilidad en la entrega.

3. How does UDP handle error detection and correction?
* UDP en sí no proporciona mecanismos para la detección o corrección de errores. En cambio, depende de protocolos o aplicaciones de capa superior para manejar estos aspectos si es necesario. Si la detección o corrección de errores es crucial para una aplicación en particular, debe implementarse en la capa de aplicación.
4. What are some common applications that use UDP?
* Algunas aplicaciones comunes donde se utilizan UDP son aplicaciones en tiempo real es decir transmisión de medios o juegos en línea. También esta en uso en las DNS ya que las consultas y respuestas de este a menudo se transmiten mediante UDP, también DHCP utiliza UDP para la comunicación entre clientes y servidores.
5. Discuss the role of ports in UDP communication.
* Los puertos en la comunicación UDP son números que identifican a qué aplicación o proceso se dirige un paquete de datos. Cada paquete UDP contiene dos números de puerto: el del emisor y el del receptor. Los puertos permiten que el sistema operativo entregue los datos a la aplicación correcta. Los puertos UDP no establecen una conexión ni garantizan la entrega de los datos, por lo que son más rápidos y ligeros que los puertos TCP. Los puertos UDP se utilizan para aplicaciones en tiempo real, como juegos, voz o video, que requieren baja latencia y pueden tolerar la pérdida de paquetes



### TCP

1. What is TCP and why is it considered a reliable protocol?
* TCP es un protocolo de capa de transporte orientado a la conexión que garantiza la entrega de datos confiable, ordenada y sin errores entre aplicaciones. Se considera confiable debido a varias características como el establecimiento de conexión ya que antes de la transmisión de datos, se establece una conexión para garantizar la comunicación entre el remitente y el receptor. Otra característica es que permite al remitente saber que los datos se entregaron correctamente. También si los datos no se confirman dentro de un período de tiempo específico, TCP los retransmite para garantizar la entrega y TCP garantiza que los datos se entreguen en el mismo orden en que se enviaron.
2. Explain the three-way handshake process used by TCP to establish a connection.
* El cliente envía un segmento TCP con el bit SYN activado al servidor, indicando su número de secuencia inicial y solicitando una conexión.
* El servidor responde con un segmento TCP con los bits SYN y ACK activados, indicando su número de secuencia inicial y reconociendo el número de secuencia del cliente.
* El cliente envía un segmento TCP con el bit ACK activado al servidor, reconociendo el número de secuencia del servidor y completando la conexión.

3. How does TCP handle flow control and congestion control?
* TCP maneja el control de flujo y el control de congestión mediante el uso de una ventana deslizante y un algoritmo de evitación de congestión. El control de flujo permite que el receptor controle la velocidad a la que el emisor envía los datos, ajustando el tamaño de la ventana deslizante según la capacidad del buffer del receptor. El control de congestión permite que el emisor controle la velocidad a la que envía los datos, ajustando el tamaño de la ventana deslizante según la estimación de la capacidad de la red.
4. Discuss the advantages and disadvantages of using TCP.

Ventajas:
* Ofrece una transmisión de datos fiable y ordenada, lo que evita la corrupción, la duplicación o la pérdida de datos.
* Proporciona mecanismos para el control de flujo y el control de congestión, lo que mejora el rendimiento y la eficiencia de la red.
* Permite la comunicación bidireccional y simultánea entre el emisor y el receptor, lo que facilita el intercambio de datos.


Desventajas:
* Es más lento y pesado que otros protocolos, como UDP, ya que requiere el establecimiento y el mantenimiento de una conexión, así como el envío y la recepción de confirmaciones.
* Es más propenso a los ataques de denegación de servicio (DoS), ya que los atacantes pueden enviar segmentos TCP falsos con el bit SYN activado para agotar los recursos del servidor.
* No es adecuado para aplicaciones en tiempo real, como juegos, voz o video, que requieren baja latencia y pueden tolerar la pérdida de paquetes.

5. What are some common applications that rely on TCP for communication?
* Algunas aplicaciones comunes que dependen de TCP para la comunicación son los servicios de correo electrónico, como SMTP, POP3 e IMAP, que utilizan TCP para enviar y recibir mensajes de forma fiable. Métodos de intercambio entre pares, como SSH y FTP, que utilizan TCP para establecer una conexión segura y transferir archivos entre dispositivos. También protocolos de aplicación web, como HTTP y HTTPS, que utilizan TCP para solicitar y entregar páginas web y otros recursos.


### Congestion Control

1. What is congestion control and why is it important in computer networks?
* El control de congestión es un mecanismo que regula la entrada de paquetes de datos en la red, evitando que se produzca un colapso por exceso de tráfico. Es importante en las redes de computadoras porque mejora el rendimiento y la eficiencia de la transmisión de datos, así como la calidad de servicio y la equidad entre los usuarios.
2. Discuss the various congestion control algorithms used in TCP.


TCP utiliza varios algoritmos de control de congestión para adaptarse a las condiciones cambiantes de la red y los algoritmos más conocidos son:
* Slow start: aumenta el tamaño de la ventana de congestión (CWND) de forma exponencial hasta alcanzar un umbral (ssthresh), para explorar la capacidad de la red
* Congestion avoidance: aumenta el tamaño de la ventana de congestión de forma lineal, sumando una constante por cada confirmación recibida, para evitar sobrecargar la red
* Fast retransmit: retransmite un segmento perdido cuando recibe tres confirmaciones duplicadas, sin esperar al temporizador, para reducir el tiempo de recuperación
* Fast recovery: reduce el tamaño de la ventana de congestión a la mitad cuando recibe tres confirmaciones duplicadas, y luego lo incrementa de forma lineal, para evitar entrar en slow start
* TCP CUBIC: utiliza una función cúbica para ajustar el tamaño de la ventana de congestión, en función del tiempo transcurrido desde el último evento de pérdida, para lograr una mayor escalabilidad y estabilidad

3. How does TCP react to congestion in the network?
* TCP reacciona a la congestión en la red reduciendo el tamaño de la ventana de congestión, lo que implica enviar menos paquetes por unidad de tiempo. De esta forma, TCP intenta aliviar la presión sobre la red y evitar la pérdida de más paquetes. TCP también retransmite los paquetes perdidos cuando recibe confirmaciones duplicadas o cuando expira el temporizador

4. What are some techniques used to measure and detect network congestion?

Algunas técnicas utilizadas para medir y detectar la congestión en la red son:
* Análisis de Pérdida de Paquetes: La pérdida de paquetes puede indicar congestión. Herramientas como traceroute y ping son comúnmente utilizadas para este propósito.
* Monitoreo del Retardo de la Red: Un aumento en los retardos puede ser un indicador de congestión. Herramientas de monitoreo de red pueden medir y analizar los retardos en tiempo real.
* Análisis de Uso de Ancho de Banda: Monitorear la utilización del ancho de banda puede revelar patrones de congestión. Herramientas como SNMP son útiles para este propósito.
* Ventanas de Congestión y Advertencias Explícitas: Algoritmos como ECN permiten a los routers marcar paquetes cuando se detecta congestión, proporcionando una señal explícita a los extremos.


### TCP and Sliding Window


1. What is the sliding window protocol and how does it improve the efficiency of data transmission?
* El protocolo de ventana deslizante es un protocolo de control de flujo para canales ruidosos. Permite al emisor enviar varios paquetes sin esperar un reconocimiento, mejorando así la eficiencia de la transmisión de datos. El protocolo utiliza una ventana deslizante para indicar el número de paquetes que el emisor puede enviar antes de recibir un reconocimiento. El protocolo también ofrece control de errores al asignar números de secuencia a los paquetes.

2. Explain the concept of window size in TCP and its impact on network performance.

* El tamaño de ventana en TCP se refiere a la cantidad máxima de datos que un remitente puede enviar antes de recibir una confirmación (acknowledgment) del receptor. Esta ventana se desliza a medida que se envían y se confirman los datos.
* Un tamaño de ventana más grande permite una mayor eficiencia en la utilización del ancho de banda, ya que más datos pueden estar en tránsito simultáneamente.
* Un tamaño de ventana más pequeño puede ser apropiado en situaciones de congestión para evitar la saturación de la red.

3. How does TCP handle acknowledgments and retransmissions using the sliding window protocol?
* Cada paquete enviado por el emisor tiene un número de secuencia y un temporizador de retransmisión. El emisor mantiene una ventana de envío que indica el rango de números de secuencia que puede enviar.
* Cada paquete recibido por el receptor tiene un número de reconocimiento que indica el siguiente número de secuencia que espera recibir. El receptor mantiene una ventana de recepción que indica el rango de números de secuencia que puede aceptar.
* Si el receptor recibe un paquete dentro de su ventana de recepción, lo almacena en el búfer y envía un reconocimiento al emisor. Si el paquete es el siguiente en orden, el receptor lo entrega a la capa superior y actualiza su ventana de recepción. Si el paquete está fuera de orden, el receptor lo almacena hasta que reciba los paquetes anteriores.
* Si el emisor recibe un reconocimiento dentro de su ventana de envío, lo interpreta como una confirmación de que el receptor ha recibido el paquete correspondiente y todos los anteriores. El emisor actualiza su ventana de envío y puede enviar más paquetes.
* Si el emisor no recibe un reconocimiento antes de que expire el temporizador de retransmisión, asume que el paquete se ha perdido o dañado y lo retransmite. El emisor también puede retransmitir un paquete si recibe tres reconocimientos duplicados, lo que indica que el receptor ha detectado un hueco en la secuencia de paquetes

4. Discuss the advantages and disadvantages of using the sliding window protocol.

Ventajas:
* Ofrece una transmisión de datos fiable y ordenada, evitando la corrupción, la duplicación o la pérdida de paquetes.
* Proporciona mecanismos para el control de flujo y el control de congestión, lo que mejora el rendimiento y la eficiencia de la red.
* Permite la comunicación bidireccional y simultánea entre el emisor y el receptor, lo que facilita el intercambio de datos.

Desventajas:
* Es más lento y pesado que otros protocolos, como UDP, ya que requiere el establecimiento y el mantenimiento de una conexión, así como el envío y la recepción de reconocimientos.
* Es más propenso a los ataques de denegación de servicio (DoS), ya que los atacantes pueden enviar paquetes falsos con el bit SYN activado para agotar los recursos del servidor.
* No es adecuado para aplicaciones en tiempo real, como juegos, voz o video, que requieren baja latencia y pueden tolerar la pérdida de paquetes.

5. What are some techniques used to optimize the sliding window protocol?

* Selective Acknowledgments (SACK).- Permite al receptor especificar qué segmentos individuales han sido recibidos correctamente, lo que mejora la eficiencia en la retransmisión de datos.
* Fast Retransmit.- TCP puede retransmitir rápidamente un segmento cuando detecta una pérdida, en lugar de esperar al temporizador de retransmisión.
* Congestion Avoidance.- Algoritmos como AIMD ayudan a evitar la congestión al ajustar la ventana de congestión en función de la detección de congestión en la red.




## Exercises

1. Consider a simple UDP-based protocol for requesting files (based somewhat loosely on the Trivial File Transport Protocol, or TFTP). The client sends an initial file request, and the server answers (if the file can be sent) with the first data packet. Client and server then continue with a stop-and-wait transmission mechanism.

    (a) Describe a scenario by which a client might request one file but get another; you may allow the client application to exit abruptly and be restarted with the same port.

    * En este escenario, el cliente envía una solicitud de archivo al servidor para obtener, por ejemplo, "arch1.txt". Sin embargo, antes de que el servidor responda, la aplicación del cliente se cierra abruptamente. Luego, el cliente se reinicia y envía otra solicitud, pero esta vez para "arch2.txt". El servidor, que aún tiene en su memoria la solicitud anterior para "arch1.txt", puede interpretar erróneamente la nueva solicitud y enviar el primer paquete de datos correspondiente a "arch1.txt" en lugar de "archivo2.txt". Esto llevaría a que el cliente reciba un archivo diferente al que solicitó.
    
    (b) Propose a change in the protocol that will make this situation much less likely.

    * Una forma de hacer que esta situación sea menos probable es implementar un mecanismo de identificación única para cada solicitud de archivo. Cuando el cliente realiza una solicitud, incluirá un identificador único en la solicitud. El servidor deberá recordar este identificador y asociarlo con el archivo correspondiente. Si el cliente se reinicia, seguirá utilizando el mismo identificador único. Cuando el servidor reciba la nueva solicitud, verificará el identificador y responderá con el archivo correcto asociado a ese identificador específico. Esto reducirá la posibilidad de que el servidor confunda solicitudes, incluso si el cliente se reinicia.

2. The sequence number field in the TCP header is 32 bits long, which is big enough to cover over 4 billion bytes of data. Even if this many bytes were never transferred over a single connection, why might the sequence number still wrap around from 2^32 − 1 to 0?

* El número de secuencia TCP es un campo de 32 bits, lo que significa que puede representar valores del 0 al 2^32−1, cubriendo más de 4 mil millones de valores diferentes. El número de secuencia se utiliza para realizar un seguimiento del flujo de bytes que se envía a través de una conexión TCP. Sin embargo, incluso si el total de datos transferidos a través de una única conexión es inferior a 4 mil millones de bytes, el número de secuencia aún podría variar de 2^32−1 a 0.

3. The Nagle algorithm, built into most TCP implementations, requires the sender to hold a partial segment’s worth of data (even if PUSHed) until either a full segment accumulates or the most recent outstanding ACK arrives.
    
    (a) Suppose the letters abcdefghi are sent, one per second, over a TCP connection with an RTT of 4.1 seconds. Draw a timeline indicating when each packet is sent and what it contains.
  * a: Sent immediately.
  * b: Held back due to the Nagle algorithm. It accumulates in the sender's buffer.
  * c: Held back as well.
  * d: The first ACK for "a" arrives, allowing the sending of "b" (Nagle condition met).
  * e: Held back.
  * f: The second ACK for "b" arrives, allowing the sending of "c."
  * g: Held back.
  * h: The third ACK for "c" arrives, allowing the sending of "d."
  * i: The fourth ACK for "d" arrives, allowing the sending of "e."

    (b) If the above were typed over a full-duplex Telnet connection, what would the user see?
    * En una conexión Telnet full-duplex, el usuario vería un retraso en la visualización de caracteres debido al algoritmo de Nagle. Los caracteres se retienen hasta que se acumula un segmento completo o llega el ACK más reciente, lo que provoca un retraso notable en la interfaz de usuario.

    (c) Suppose that mouse position changes are being sent over the connection. Assuming that multiple position changes are sent each RTT, how would a user perceive the mouse motion with and without the Nagle algorithm?
    * Con el algoritmo de Nagle los cambios de posición del mouse se retendrían hasta que se acumula un segmento completo o hasta que llega el ACK más reciente. Esto podría resultar en una percepción más lenta de los cambios de posición del mouse, ya que se agruparían y se enviarían en intervalos más grandes.
Sin el algoritmo de Nagle cada cambio de posición del mouse se enviaría de inmediato sin esperar la acumulación de un segmento completo. Esto podría llevar a una percepción más rápida y suave de los movimientos del mouse, ya que se enviarían tan pronto como estén disponibles.


5. Suppose that between A and B there is a router R. The A–R bandwidth is infinite (that is, packets are not delayed), but the R–B link introduces a bandwidth delay of 1 packet per second (that is, 2 packets take 2 seconds, etc.). Acknowledgments from B to R, though, are sent instantaneously. A sends data to B over a TCP connection, using slow start but with an arbitrarily large window size. R has a queue size of one, in addition to the packet it is sending. At each second, the sender first processes any arriving ACKs and then responds to any timeouts.

    (a) Assuming a fixed TimeOut period of 2 seconds, what is sent and received for T = 0, 1, . . . , 6 seconds? Is the link ever idle due to timeouts?
      * T = 0 segundos: A envía el primer paquete.
      * T = 1 segundo: El paquete llega a B. B envía un ACK instantáneo.
      * T = 2 segundos: A no ha recibido el ACK, asume la pérdida y reenvía el paquete.
      * T = 3 segundos: El paquete original y el reenviado llegan a B. B envía un ACK instantáneo.
      * T = 4 segundos: A recibe ambos ACK. No se envía nada.
      * T = 5 segundos: A no ha recibido el ACK del paquete reenviado, asume la pérdida y reenvía el paquete.
      * T = 6 segundos: El paquete original y el segundo reenviado llegan a B. B envía un ACK instantáneo.

    (b) What changes if TimeOut is 3 seconds instead?
    * Las retransmisiones se retrasarán hasta que se agote el período de TimeOut. Esto afectará el rendimiento y podría hacer que la conexión sea menos sensible a las pérdidas. En términos de datos enviados y recibidos, el escenario sería similar al caso anterior, pero las retransmisiones ocurrirían en intervalos de 3 segundos en lugar de 2.
