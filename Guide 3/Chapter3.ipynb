## Switching Basics

### Questions:
1. What is switching in computer networking?
* El switching es una técnica utilizada en redes de computadoras para conectar dispositivos y permitir la transferencia de datos entre ellos. En términos simples, un switch es un dispositivo que conecta varios dispositivos en una red y permite que los datos se transfieran entre ellos.
2. Explain the difference between circuit switching and packet switching.
* Circuit switching es una técnica en la que se establece un circuito dedicado entre dos dispositivos antes de que se transmitan los datos. En otras palabras, se reserva un camino dedicado para la comunicación entre los dispositivos. Por otro lado, en el packet switching, los datos se dividen en paquetes y se envían a través de la red. Cada paquete se enruta de forma independiente y puede seguir diferentes caminos para llegar a su destino.
3. What are the advantages of packet switching over circuit switching?
* Packet switching tiene varias ventajas sobre circuit switching, como una mayor eficiencia en el uso del ancho de banda, una mayor flexibilidad y escalabilidad, y una menor latencia. Además, el packet switching es más adecuado para aplicaciones que no requieren una conexión dedicada, como la transmisión de datos a través de Internet.
4. What is a switch? How does it differ from a hub?
* Un switch es un dispositivo de red que se utiliza para conectar varios dispositivos en una red y permitir que los datos se transfieran entre ellos. A diferencia de un hub, un switch es capaz de identificar la dirección MAC de cada dispositivo conectado a él y enviar los datos solo al dispositivo de destino. Esto reduce la cantidad de tráfico en la red y mejora el rendimiento.
5. Describe the process of switching a packet within a switch.
* Cuando un paquete llega a un switch, el switch examina la dirección MAC de destino del paquete y busca en su tabla de direcciones para determinar el puerto de salida adecuado. Si el puerto de salida se encuentra en la misma red que el puerto de entrada, el switch envía el paquete directamente al dispositivo de destino. Si el puerto de salida se encuentra en una red diferente, el switch envía el paquete al router adecuado para su reenvío.
6. What is store-and-forward switching? How does it ensure data integrity?
* El switching store-and-forward es una técnica en la que un switch recibe todo el paquete antes de reenviarlo al dispositivo de destino. El switch verifica la integridad del paquete antes de reenviarlo, lo que garantiza que los datos se transmitan sin errores. Si se detecta un error, el paquete se descarta y no se envía al dispositivo de destino.

### Exercises:
1. Compare and contrast the performance characteristics of circuit switching and packet switching.
* El circuit switching proporciona un ancho de banda garantizado y una latencia baja, pero es menos eficiente en el uso del ancho de banda y no es escalable. El packet switching es más eficiente en el uso del ancho de banda y es altamente escalable, pero puede tener una latencia más alta y no garantiza un ancho de banda dedicado.
2. Research and list three different types of switches commonly used in computer networks.
* Switches modulares: estos switches permiten la personalización de la red mediante la adición o eliminación de módulos según sea necesario. Los módulos pueden incluir puertos Ethernet, puertos de fibra óptica, etc. Algunos ejemplos de switches modulares incluyen el Cisco Catalyst 6500 y el HP ProCurve 5400.

* Switches de agregación: estos switches se utilizan para conectar varios switches de acceso y crear una red de mayor tamaño. Los switches de agregación también pueden proporcionar funciones de seguridad y administración de red. Algunos ejemplos de switches de agregación incluyen el Cisco Nexus 7000 y el HP ProCurve 8200.

* Switches de centro de datos: estos switches se utilizan en centros de datos para proporcionar conectividad de alta velocidad y baja latencia entre servidores y dispositivos de almacenamiento. Los switches de centro de datos también pueden proporcionar funciones de virtualización y automatización de red. Algunos ejemplos de switches de centro de datos incluyen el Cisco Nexus 9000 y el HP FlexFabric 12900.

## Switched Ethernet

### Questions:
1. What is Ethernet? How does it work?
* Ethernet es una tecnología de red que se utiliza para conectar dispositivos en una red local. Funciona mediante el envío de paquetes de datos entre dispositivos a través de un cableado físico. Los dispositivos en una red Ethernet se comunican entre sí utilizando direcciones MAC únicas.

2. Explain the concept of a collision domain in Ethernet.
* Un dominio de colisión es un segmento de red en el que los paquetes de datos pueden colisionar entre sí. En una red Ethernet, un dominio de colisión se produce cuando dos o más dispositivos transmiten datos al mismo tiempo. Esto puede causar una disminución en el rendimiento de la red. Los hubs y los repetidores amplían los dominios de colisión, mientras que los switches y los routers los reducen.

3. What are learning bridges?
* Los puentes de aprendizaje son dispositivos de red que se utilizan para conectar varias redes de área local (LAN) y permitir que los datos se transfieran entre ellas. Los puentes de aprendizaje utilizan tablas de direcciones MAC para determinar la ubicación de los dispositivos en la red y enviar los datos solo al dispositivo de destino. Esto reduce la cantidad de tráfico en la red y mejora el rendimiento.

4. Explain the Spanning Tree Algorithm
* El algoritmo Spanning Tree es un algoritmo utilizado para evitar bucles en la topología de la red. El algoritmo funciona mediante la selección de un árbol de expansión mínimo en la red.

5. What are VLANs?
* Las VLAN (Virtual LAN) son una técnica utilizada en redes de computadoras para crear redes lógicas independientes dentro de una red física. Las VLAN se utilizan para segmentar una red en subredes más pequeñas y mejorar la seguridad y el rendimiento de la red.

## Internet (IP)

### Questions:
1. What is the Internet Protocol (IP)? How does it work?
*  El Protocolo de Internet (IP) es un protocolo de comunicaciones de red que se utiliza para enviar y recibir datos a través de Internet. IP es responsable de enrutar los paquetes de datos desde la fuente hasta el destino correcto. Cuando se envía un paquete de datos, se divide en trozos más pequeños llamados paquetes. Cada paquete contiene información de encabezado que incluye la dirección IP del remitente y del destinatario. Los routers utilizan esta información para enrutar los paquetes a través de la red hasta que llegan a su destino final.
2. Explain the difference between IPv4 and IPv6.
* IPv4 e IPv6 son dos versiones diferentes del Protocolo de Internet. IPv4 es la versión anterior y utiliza direcciones de 32 bits, lo que limita el número de direcciones IP únicas disponibles. IPv6 es la versión más reciente y utiliza direcciones de 128 bits, lo que permite un número mucho mayor de direcciones IP únicas. IPv6 también ofrece mejoras en la seguridad y la eficiencia en comparación con IPv4.
3. What is an IP address? How is it structured?
* Una dirección IP es un identificador único que se asigna a cada dispositivo conectado a una red que utiliza el Protocolo de Internet. Las direcciones IP se utilizan para enrutar los paquetes de datos a través de la red hasta su destino final. Una dirección IP consta de dos partes: la identificación de la red y la identificación del host. La identificación de la red identifica la red a la que está conectado el dispositivo, mientras que la identificación del host identifica el dispositivo específico en esa red. Las direcciones IP se expresan generalmente como una serie de cuatro números separados por puntos, como 192.168.1.1.
4. What is the purpose of subnetting in IP addressing?
* El subnetting es una técnica utilizada para dividir una red en subredes más pequeñas. Esto se hace para mejorar la eficiencia de la red y reducir el tráfico de red no deseado. El subnetting se utiliza para crear subredes que tienen su propia dirección de red y rango de direcciones IP. Esto permite que los dispositivos en la subred se comuniquen directamente entre sí sin tener que pasar por un router.
5. Describe the process of IP packet forwarding.
* El forwarding de paquetes IP es el proceso de enviar paquetes de datos desde la fuente hasta el destino correcto a través de la red. Cuando se envía un paquete de datos, se divide en trozos más pequeños llamados paquetes. Cada paquete contiene información de encabezado que incluye la dirección IP del remitente y del destinatario. Los routers utilizan esta información para enrutar los paquetes a través de la red hasta que llegan a su destino final.

6. What is the role of the Internet Control Message Protocol (ICMP) in IP?
* El Protocolo de Control de Mensajes de Internet (ICMP) es un protocolo utilizado para enviar mensajes de error y control a otros dispositivos de red. ICMP se utiliza para informar a los dispositivos de red sobre problemas de conectividad, como paquetes perdidos o inaccesibles.

## Routing

### Questions:
1. What is routing in computer networking?
* El routing es el proceso de seleccionar la mejor ruta para que los datos viajen de una red a otra. Implica el reenvío de paquetes de datos desde una fuente a un destino a través de múltiples redes.

2. Explain the difference between static routing and dynamic routing.
* El routing estático es un proceso manual en el que el administrador de la red ingresa manualmente las rutas en la tabla de routing, mientras que el routing dinámico es un proceso automatizado en el que las rutas se encuentran automáticamente según los cambios en la red.

3. What is a routing table? How is it used in the routing process?
* Una tabla de routing es una base de datos que realiza un seguimiento de las rutas y las utiliza para determinar la mejor manera de reenviar el tráfico. Contiene información sobre la topología de la red inmediatamente a su alrededor.

4. What is the purpose of a default gateway in routing?
* Una puerta de enlace predeterminada es una ruta a la que se pasa información cuando el dispositivo no sabe dónde se encuentra el destino. Se utiliza cuando no hay información de routing disponible sobre el destino. Es un dispositivo importante para el reenvío de datos y el routing de los datos en la otra red.

5. Describe the process of routing a packet from source to destination.
* Cuando se envía un paquete desde una fuente a un destino, primero se envía al router más cercano. El router examina el encabezado IP del paquete para determinar la dirección IP de destino. Luego utiliza su tabla de forwarding para determinar la mejor ruta hacia el destino. El paquete se envía al siguiente router a lo largo de la ruta hasta que llega a su destino final.
